```python
top10 = df.sort_values(
    by='Population',
    ascending=False
).head(10)

top10
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x='Country Name',
    y='Population',
    data=top10
)

plt.xticks(rotation=45)

plt.title("Top 10 Most Populated Countries")

plt.show()
```

```python
growth = df.sort_values(
    by='Population Growth Rate',
    ascending=False
).head(10)

growth[['Country Name', 'Population Growth Rate']]
```

```python
df['Population Density'] = df['Population'] / df['Land Area']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))

sns.scatterplot(
    x='Birth Rate',
    y='Death Rate',
    data=df
)

plt.title("Birth Rate vs Death Rate")

plt.show()
```

```python
migration = df.sort_values(
    by='Migration Rate',
    ascending=False
).head(10)

migration[['Country Name', 'Migration Rate']]
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    x='Area',
    y='Population',
    data=df
)

plt.title("Area vs Population")

plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')

plt.figure(figsize=(12,8))

sns.heatmap(
    numeric_df.corr(),
    annot=True
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
negative_migration = df[df['Migration Rate'] < 0]

negative_migration[['Country Name', 'Migration Rate']].head(10)
```

```python
plt.figure(figsize=(10,6))

sns.regplot(
    x='Birth Rate',
    y='Population Growth Rate',
    data=df
)

plt.title("Birth Rate vs Population Growth")

plt.show()
```
