```python
top10 = df.sort_values(
    by='Population',
    ascending=False
).head(10)

top10
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x='Country Name',
    y='Population',
    data=top10
)

plt.xticks(rotation=45)

plt.title("Top 10 Most Populated Countries")

plt.show()
```

```python
growth = df.sort_values(
    by='Population Growth Rate',
    ascending=False
).head(10)

growth[['Country Name', 'Population Growth Rate']]
```

```python
df['Population Density'] = df['Population'] / df['Land Area']
```
